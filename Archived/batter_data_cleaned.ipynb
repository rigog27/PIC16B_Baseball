```python
import pandas as pd
```

```python
df = pd.read_csv('dodgers_data.csv')
df
```

```python
df_fb = pd.read_csv('dodgers_fb.csv')
df_fb
```

```python
df_os = pd.read_csv('dodgers_os.csv')
df_os
```

```python
df_sl = pd.read_csv('dodgers_sl.csv')
df_cb = pd.read_csv('dodgers_cb.csv')
```

```python
df_fb['pitch_type'] = 'fastball'
df_fb = df_fb[['player_id', 'player_name', 'pitch_type', 'pitches', 'pitch_percent', 'ba', 'slg', 'xba', 'obp' ]] 
df_fb 
```

```python
df_os['pitch_type'] = 'off-speed'
df_os = df_os[['player_id', 'player_name', 'pitch_type', 'pitches', 'pitch_percent', 'ba', 'slg', 'xba', 'obp' ]] 
df_os
```

```python
df_sl['pitch_type'] = 'slider'
df_sl = df_sl[['player_id', 'player_name', 'pitch_type', 'pitches', 'pitch_percent', 'ba', 'slg', 'xba', 'obp' ]] 
df_cb['pitch_type'] = 'curveball'
df_cb = df_cb[['player_id', 'player_name', 'pitch_type', 'pitches', 'pitch_percent', 'ba', 'slg', 'xba', 'obp' ]] 
```

```python
df_all = pd.concat([df_fb, df_os, df_sl, df_cb], ignore_index=True)

#Sort by pitch_type then pitches
df_all = df_all.sort_values(['pitch_type', 'pitches'], ascending=[True,False]).reset_index(drop=True)

df_all.head(60)
```

```python
#Pivot into each player being a row
df_wide = (
    df_all.pivot_table(
        index=['player_id', 'player_name'],
        columns='pitch_type',
        values=['ba', 'slg', 'xba', 'obp', 'pitches', 'pitch_percent']
    )
)

#Change column names
df_wide.columns = [f'{stat}_{ptype}' for stat, ptype in df_wide.columns]
df_wide = df_wide.reset_index()

#Sort players alphabetically
df_wide = df_wide.sort_values('player_name').reset_index(drop=True)
#Drop players with NaN values (as they are no longer with the team)
df_wide = df_wide.dropna()
df_wide
```
