```python
import pandas as pd
```

```python
# read csv into pandas df
dodger_pitches = pd.read_csv("data/dodgers_pitchers_vs_bluejays.csv")

# get dataframe with player ids
ids = pd.read_csv("data/mlbids.csv")
```

```python
# take only wanted values that can be measured during the current play
dodger_pitches = dodger_pitches[['pitch_type', 'player_name', 'batter', 
                                 'balls', 'strikes', 'on_3b', 'on_2b', 
                                 'on_1b', 'pitch_name', 'fld_score', 
                                 'bat_score']]
```

```python
# add a count column
#dodger_pitches['count'] = str(dodger_pitches['balls'])
```

```python
# replace player id on bases 1,2,3 to true or false if base is occupied
dodger_pitches[['on_3b', 'on_2b', 'on_1b']] = dodger_pitches[['on_3b', 'on_2b', 'on_1b']].notna()
```

```python
# get player ids and names
merged = dodger_pitches.merge(ids, left_on='batter', right_on='MLBID', how='left')
```
